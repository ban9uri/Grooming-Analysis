```python
from konlpy.tag import Okt
from konlpy.tag import Komoran
from konlpy.tag import Kkma
okt = Okt()
komoran = Komoran()
tagger = Komoran()
kkma = Kkma()
```

```python
###제품명

df=pd.read_table('/content/drive/MyDrive/남성 스킨케어 n=120.txt')
#print(df)

twitter.add_dictionary('우르오스', 'Noun')
twitter.add_dictionary('플루이드', 'Noun')
twitter.add_dictionary('모이스처', 'Noun')
twitter.add_dictionary('쉐이브', 'Noun')
twitter.add_dictionary('데일리', 'Noun')
twitter.add_dictionary('시카', 'Noun')
twitter.add_dictionary('알로에', 'Noun')
twitter.add_dictionary('프레쉬', 'Noun')
twitter.add_dictionary('레드', 'Noun')
twitter.add_dictionary('약산성', 'Noun')
twitter.add_dictionary('중건성', 'Noun')
twitter.add_dictionary('리페어', 'Noun')
twitter.add_dictionary('클렌저', 'Noun')
twitter.add_dictionary('페이셜', 'Noun')

a=[]

stopwords=['방기', '라이', '텐드']
for sentence in (df['제품명']):
  a.append(sentence)
tagged_sentences=[twitter.pos(sent, norm=True, stem=True) for sent in a]
#print(tagged_sentences)

word_list = []
for sent in tagged_sentences:   
    for word, tag in sent:
      #if word not in stopwords:
       if len(word)>1 and word not in stopwords:
        if tag in ['Noun']:
            word_list.append(word)

word_counts = Counter(word_list)
print(word_counts)
type(word_counts)

cloud = WordCloud(width=900, height=600,
      font_path='/content/drive/MyDrive/08서울남산체 B.ttf',
      background_color='white')
cloud = cloud.fit_words(word_counts)
plt.figure(figsize=(15, 20))
plt.axis('off')
plt.imshow(cloud)
plt.show()
```

```python
###마케팅문구

df=pd.read_table('/content/drive/MyDrive/마케팅 문구.txt')
#print(df)

stopwords=['있다']

a=[]

for sentence in (df['11111111111111111111111111. ']):
  a.append(sentence)
tagged_sentences=[okt.pos(sent, norm=True, stem=True) for sent in a]
#print(tagged_sentences)

word_list = []
for sent in tagged_sentences:    
    for word, tag in sent:
      if word not in stopwords:
        if tag in ['Adjective']:
            word_list.append(word)

word_counts = Counter(word_list)
print(word_counts)

cloud = WordCloud(width=900, height=600, 
      font_path='/content/drive/MyDrive/08서울남산체 B.ttf',
      background_color='white')
cloud = cloud.fit_words(word_counts)
plt.figure(figsize=(15, 20))
plt.axis('off')
plt.imshow(cloud)
plt.show()
```
