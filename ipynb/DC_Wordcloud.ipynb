```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import numpy as np
from collections import Counter
```

```python
from konlpy.tag import Okt
from konlpy.tag import Komoran
komoran = Komoran()
okt = Okt()
#print(okt.morphs('인터넷에서 쓰이는 용어도 분석이 되나욬ㅋㅋ', norm=True))
tagger = Komoran()
```

```python
df=pd.read_csv('/content/drive/MyDrive/dc_negative.csv')

a=[]

for sentence in (df['comment']):
  a.append(sentence)
tagged_sentences=[tagger.pos(sent) for sent in a]

adverb_list = []
for sent in tagged_sentences:    
    for word, tag in sent:
        if tag in ['VA']:
            adverb_list.append(word)

adverb_counts = Counter(adverb_list)

cloud = WordCloud(width=900, height=600, 
      font_path='/content/drive/MyDrive/08서울남산체 B.ttf',
      background_color='white')
cloud = cloud.fit_words(adverb_counts)
plt.figure(figsize=(15, 20))
plt.axis('off')
plt.imshow(cloud)
plt.show()
```
