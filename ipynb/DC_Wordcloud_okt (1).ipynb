```python
###마케팅문구

df=pd.read_table('/content/drive/MyDrive/마케팅 문구.txt')
#print(df)

a=[]

for sentence in (df['11111111111111111111111111. ']):
  a.append(sentence)
#tagged_sentences=[tagger.pos(sent) for sent in a]

tagged_sentences=[okt.pos(sent,norm=True, stem=True) for sent in a]


word_list = []
for sent in tagged_sentences:    
    for word, tag in sent:
        if tag in ['Adjective']:
            word_list.append(word)

word_counts = Counter(word_list)
print(word_counts)

cloud = WordCloud(width=900, height=600, 
      font_path='/content/drive/MyDrive/08서울남산체 B.ttf',
      background_color='white')
cloud = cloud.fit_words(word_counts)
plt.figure(figsize=(15, 20))
plt.axis('off')
plt.imshow(cloud)
plt.show()
```

```python
###제품명

df=pd.read_table('/content/drive/MyDrive/남성 스킨케어 제품명 n=120.txt')
#print(df)

kkma = Kkma()



stopwords = ['개', '용','쉬']

a=[]

for sentence in (df['제품명']):
  a.append(sentence)
tagged_sentences=[okt.pos(sent, norm=True, stem=True) for sent in a]
#print(tagged_sentences)

word_list = []
for sent in tagged_sentences:   
    for word, tag in sent:
      if word not in stopwords:
        if tag in ['Noun']:
            word_list.append(word)

word_counts = Counter(word_list)
print(word_counts)

cloud = WordCloud(width=900, height=600, 
      font_path='/content/drive/MyDrive/08서울남산체 B.ttf',
      background_color='white')
cloud = cloud.fit_words(word_counts)
plt.figure(figsize=(15, 20))
plt.axis('off')
plt.imshow(cloud)
plt.show()
```

```python
#트위터 형용사 Wordcloud

import numpy as np
from PIL import Image

male_mask=np.array(Image.open('/content/drive/MyDrive/female1.png'))
male_mask

for i in range(len(male_mask)):
    for j in range(len(male_mask[i])):
        if np.all(male_mask[i][j])==0:
            male_mask[i][j] =255

stop_words = '입니다 같다 있다 없다 그렇다 이다 아니다 확실하다 어떻다 남자 관리 이렇다 야하다'
stop_words=stop_words.split(' ')


df=pd.read_csv('/content/drive/MyDrive/twitter_crawling.csv')

a=[]

for sentence in (df['Tweet']):

  a.append(sentence)
tagged_sentences=[okt.pos(sent, norm=True, stem=True) for sent in a]

adverb_list = []
for sent in tagged_sentences:    
    for word, tag in sent:
      if word not in stop_words:
        if tag in ['Adjective']:
            adverb_list.append(word)

adverb_counts = Counter(adverb_list)
print(adverb_counts)

cloud = WordCloud(mask=male_mask, width=500, height=500, 
      font_path='/content/drive/MyDrive/08서울남산체 B.ttf', colormap='autumn', 
      background_color='white')
cloud = cloud.fit_words(adverb_counts)
plt.figure(figsize=(25, 25))
plt.axis('off')
plt.imshow(cloud)
plt.show()
plt.savefig('test.png')
```
