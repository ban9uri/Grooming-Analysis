```python
from pororo import Pororo
```

```python
Pororo.available_tasks()
```

```python
sa_shop = Pororo(task='sentiment', model='brainbert.base.ko.shopping', lang='ko')
sa = Pororo(task="sentiment", model="brainbert.base.ko.nsmc", lang="ko")
```

```python
import pandas as pd
df = pd.read_excel("youtube crawling.xlsx")
```

```python
df
```

```python
comment = []
for i in range(len(df.comment)):
    if len(df.comment[i]) <= 512:
        comment.append(df.comment[i])

sentiment = []
for i in range(len(comment)):
    sentiment.append(sa(comment[i]))

positive = []
for i in range(len(comment)):
    positive.append(sa(comment[i], show_probs=True)['positive'])

negative = []
for i in range(len(comment)):
    negative.append(sa(comment[i], show_probs=True)['negative'])

df_new = pd.DataFrame()
df_new["comment"] = comment
df_new["sentiment"] = sentiment
df_new["positive"] = positive
df_new["negative"] = negative

for_df = []

for i in range(len(df_new["comment"])):
    if df_new.iloc[i][1] == 'Positive':
        if df_new.iloc[i][2] >= 0.85:
            for_df.append(df_new.iloc[i])
    elif df_new.iloc[i][1] == 'Negative':
        if df_new.iloc[i][3] >= 0.85:
            for_df.append(df_new.iloc[i])

print("필터링 된 댓글의 수:", len(for_df))

df_new2 = pd.DataFrame(for_df)
df_new2
```

```python
condition_ps = (df_new2.sentiment == "Positive")
condition_ng = (df_new2.sentiment == "Negative")
```

```python
df_positive = df_new2[condition_ps]
df_negative = df_new2[condition_ng]
```

```python
df_positive.to_csv('youtube_positive.csv', encoding='utf-8-sig') 
df_negative.to_csv('youtube_negative.csv', encoding='utf-8-sig') 
```
