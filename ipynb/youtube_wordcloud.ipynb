```python
%matplotlib inline
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import numpy as np
from collections import Counter
```

```python
from konlpy.tag import Okt
from konlpy.tag import Komoran
komoran = Komoran()
okt = Okt()
tagger = Komoran()
```

```python
# 유튜브 부정
 df=pd.read_csv('/content/drive/MyDrive/youtube_negative_2.csv')

a=[]

for sentence in (df['comment']):
  a.append(sentence)
tagged_sentences=[okt.pos(sent) for sent in a]

word_list = []
for sent in tagged_sentences:    
    for word, tag in sent:
        if tag in ['Noun']:
            word_list.append(word)

word_counts = Counter(word_list)

cloud = WordCloud(width=900, height=600, 
      font_path='/content/drive/MyDrive/_DataJournalism_2022/Week8/data/08서울남산체 B.ttf',
      background_color='white')
cloud = cloud.fit_words(word_counts)
plt.figure(figsize=(15, 20))
plt.axis('off')
plt.imshow(cloud)
plt.show()
```

```python
# 별점 긍정
import nltk
from nltk.corpus import stopwords 
from nltk.tokenize import word_tokenize 


df=pd.read_csv('/content/drive/MyDrive/star45.csv')

a=[]

for sentence in (df['comment']):
  a.append(sentence)
tagged_sentences=[okt.pos(sent, norm=True, stem=True) for sent in a]
tagged_sentences

word_list = []
for sent in tagged_sentences:    
    for word, tag in sent:
        if tag in ['Adjective']:
            word_list.append(word)


stop_words = '입니다 같다 좋다 있다 없다 그렇다 이다 아니다 확실하다 괜찮다 어떻다 야하다 '
stop_words=stop_words.split(' ')

result = []
for w in word_list: 
    if w not in stop_words: 
        result.append(w) 

word_counts = Counter(result)

cloud = WordCloud(width=900, height=600, 
      font_path='/content/drive/MyDrive/_DataJournalism_2022/Week8/data/08서울남산체 B.ttf',
      background_color='white')
cloud = cloud.fit_words(word_counts)
plt.figure(figsize=(15, 20))
plt.axis('off')
plt.imshow(cloud)
plt.show()
```
