```python
# 레오제이 그래프
import pandas as pd
from google.colab import drive
drive.mount('/content/drive')
data = pd.read_excel("/content/drive/MyDrive/LeoJ_result.xlsx")

%matplotlib inline
import matplotlib.pyplot as plt
x = data.loc[:,'날짜']
y = data.loc[:,'구독자 수']

plt.plot(x,y)

current_values = plt.gca().get_yticks()
plt.gca().set_yticklabels(['{:.0f}'.format(x) for x in current_values])

plt.show()
```

```python
# 스완 그래프
import pandas as pd
from google.colab import drive
drive.mount('/content/drive')
data = pd.read_excel("/content/drive/MyDrive/Swan_result.xlsx")

%matplotlib inline
import matplotlib.pyplot as plt

x = data.loc[:,'날짜']
y = data.loc[:,'구독자 수']

plt.plot(x,y)

# axes.xaxis.set_major_locator(MaxNLocator(3)) 
current_values = plt.gca().get_yticks()
plt.gca().set_yticklabels(['{:.0f}'.format(x) for x in current_values])
plt.xticks(x)
plt.show()
```
